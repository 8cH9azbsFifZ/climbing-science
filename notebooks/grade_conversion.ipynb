# Grade Conversion

Demonstrates converting between climbing grading systems:
- **Route**: UIAA ↔ French ↔ YDS
- **Boulder**: Fontainebleau ↔ V-Scale

Based on the IRCRA comparative grading table (Draper et al. 2015).

```python
from climbing_science.grades import convert, parse, compare, from_index, all_grades
```

## Route Grade Conversion

```python
convert("VII+", "UIAA", "French")
```

```python
convert("7a", "French", "YDS")
```

```python
convert("5.12a", "YDS", "UIAA")
```

## Boulder Grade Conversion

```python
convert("7A", "Font", "V-Scale")
```

```python
convert("V5", "V-Scale", "Font")
```

## Auto-Detection

`parse()` detects the grading system from the grade string automatically.

```python
for grade_str in ["6b+", "5.11a", "VII", "7A+", "V4"]:
    g = parse(grade_str)
    print(f'  "{grade_str}" -> system={g.system.value}, IRCRA index={g.difficulty_index}')
```

## Cross-System Comparison

Compare grades from different systems via their IRCRA difficulty index.

```python
for a, b in [("VII+", "6b+"), ("5.12a", "7a+"), ("V5", "6C")]:
    result = compare(a, b)
    symbol = "=" if result == 0 else (">" if result > 0 else "<")
    print(f"  {a} {symbol} {b}")
```

## Reverse Lookup: IRCRA Index → Grade

Find the nearest grade for a given difficulty index.

```python
for idx in [15, 20, 25, 30]:
    uiaa = from_index(idx, "UIAA")
    french = from_index(idx, "French")
    yds = from_index(idx, "YDS")
    print(f"  IRCRA {idx}: UIAA={uiaa}, French={french}, YDS={yds}")
```

## Full Grade Table
